```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleanfpl_a.csv')
```

```python
df.head(8)
```

### Looking at some correlations between the variables

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
plt.figure(figsize=(14,14))
sns.heatmap(df[["xValue/90","Value/90","Price","SiB","Goals"]].corr(), cmap=cmap, vmax=.8, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
sns.regplot(df['xValue/90'],df['Value/90'])
plt.show()
#Shows people are more likely to extremely overperform on xValue then underperfrom. 
#So if someone has good xValue/90 they can't really be a bad pick as long as they get mins.
```

```python
sns.regplot(df['xValue/App'],df['Value/App'])
plt.show()
```

```python
sns.regplot(df['Price'],df['Value/90'])
plt.show()
```

```python
sns.regplot(df['Price'],df['xValue/90'])
plt.show()
```

```python
sns.regplot(df['Price'],df['xPoints'])
plt.show()
```

```python
sns.regplot(df['SiB'],df['Goals'])
plt.show()
```
